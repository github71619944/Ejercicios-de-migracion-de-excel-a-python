```python

import pandas as pd
import numpy as np
```

```python
# leer archivo xlsx en un dataframe
df=pd.read_excel('datos.xlsx')
```

```python
df.head(70)
```

```python
#Calcular el Total General de Sueldo Básico de los Trabajadores (Total de la Planilla)
# sumar valores de una columna
Sueldo_Básico = df['SUELDO\nBASICO'].sum()

print('el Total General de Sueldo Básico de los Trabajadores es: {}'.format(Sueldo_Básico))
```

```python
#Calcular el Total de Sueldo Básico de los Trabajadores
#Sueldo Básico Máximo
Sueldo_Básico_Maximo = df['SUELDO\nBASICO'].max()
#Sueldo Básico Mínimo
Sueldo_Básico_Minimo = df['SUELDO\nBASICO'].min()

print('El Sueldo Básico Máximo es: {}'.format(Sueldo_Básico_Maximo))
print('El Sueldo Básico Mínimo es: {}'.format(Sueldo_Básico_Minimo))
```

```python
df.describe()
```

```python
#Calcular el Total de Trabajadores

# cuente el número de trabajadores estables y no estables.
conteo_estable = df['CONDICION'].value_counts()['ESTABLE']
conteo_contratado = df['CONDICION'].value_counts()['CONTRATADO']

# imprimir resultados
print('Trabajadores estables: {}'.format(conteo_estable))
print('Trabajadores contratados: {}'.format(conteo_contratado))
```

```python
#Calcular el Cantidad de Trabajadores por Ocupación

# cuente el número de trabajadores segun su tipo 

N_ABOGADO = df['OCUPACION'].value_counts()['ABOGADO']
N_ADMINISTRADOR = df['OCUPACION'].value_counts()['ADMINISTRADOR']
N_ALMACEN = df['OCUPACION'].value_counts()['ALMACEN']
N_ANALISTA = df['OCUPACION'].value_counts()['ANALISTA']
N_CONTADOR = df['OCUPACION'].value_counts()['CONTADOR']
N_EMPLEADO = df['OCUPACION'].value_counts()['EMPLEADO']
N_GERENTE = df['OCUPACION'].value_counts()['GERENTE']
N_MARKETING = df['OCUPACION'].value_counts()['MARKETING']
N_OBRERO = df['OCUPACION'].value_counts()['OBRERO']
N_OPERADOR = df['OCUPACION'].value_counts()['OPERADOR']
N_PROGRAMADOR = df['OCUPACION'].value_counts()['PROGRAMADOR']
N_SECRETARIA = df['OCUPACION'].value_counts()['SECRETARIA']
N_SERVICIOS = df['OCUPACION'].value_counts()['SERVICIOS']
N_TECNICO_E = df['OCUPACION'].value_counts()['TECNICO E.']


# imprimir resultados
print('NUMERO DE ABOGADOS : {}'.format(N_ABOGADO))
print('NUMERO DE ADMINISTRADORES : {}'.format(N_ADMINISTRADOR))
print('NUMERO DE ALMACENEROS : {}'.format(N_ALMACEN))
print('NUMERO DE ANALISTAS : {}'.format(N_ANALISTA))
print('NUMERO DE CONTADORES : {}'.format(N_CONTADOR))
print('NUMERO DE EMPLEADOS : {}'.format(N_EMPLEADO))
print('NUMERO DE GERENTES : {}'.format(N_GERENTE))
print('NUMERO DE MARKETING : {}'.format(N_MARKETING))
print('NUMERO DE OBREROS : {}'.format(N_OBRERO))
print('NUMERO DE OPERADORES : {}'.format(N_OPERADOR))
print('NUMERO DE PROGRAMADORES : {}'.format(N_PROGRAMADOR))
print('NUMERO DE SECRETARIAS : {}'.format(N_SECRETARIA))
print('NUMERO DE SERVICIOS : {}'.format(N_SERVICIOS))
print('NUMERO DE TECNICOS : {}'.format(N_TECNICO_E))
```
